```python
import sympy as sp

x, y = sp.symbols('x y')

# Defining the function
f = sp.exp(x) * sp.sin(y) + y**3

df_dx = sp.diff(f, x)  # with respect to x
df_dy = sp.diff(f, y)  # with respect to y

df_dx, df_dy
```

```python
import sympy as sp

x, y = sp.symbols('x y')
g = x**2 * y + x * y**2

grad_g_x = sp.diff(g, x)  # with respect to x
grad_g_y = sp.diff(g, y)  # with respect to y
gradient = (grad_g_x, grad_g_y)

# gradient at (1, -1)
gradient_at_point = [expr.subs({x: 1, y: -1}) for expr in gradient]

magnitude = sp.sqrt(gradient_at_point[0]**2 + gradient_at_point[1]**2)

gradient, gradient_at_point, magnitude
```

```python
import sympy as sp

x, y = sp.symbols('x y')

h = sp.ln(x**2 + y**2)

# second partial derivatives
h_xx = sp.diff(h, x, x) 
h_yy = sp.diff(h, y, y) 
h_xy = sp.diff(h, x, y)

h_xx, h_yy, h_xy
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# symbols
x, y = sp.symbols('x y')

# function
j = x**3 - 3*x*y + y**3

j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

# Plotting
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title(r'Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')

# my chosen function
m = (x**3 + y**3) / (1 + x**2 + y**2)

m_func = sp.lambdify((x, y), m, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = m_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='coolwarm')
plt.colorbar()
plt.title(r'Contour plot of $m(x, y) = \frac{x^3 + y^3}{1 + x^2 + y^2}$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.show()
```

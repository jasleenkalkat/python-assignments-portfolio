```python
import numpy as np

def normal_density(mean, variance, x):
  
    coefficient = 1 / np.sqrt(2 * np.pi * variance)
    exponent = -((x - mean) ** 2) / (2 * variance)
    return coefficient * np.exp(exponent)

import matplotlib.pyplot as plt

def plot_normal_density():
    x_values = np.linspace(150, 200, 500)
    params = [(171, 7.1), (160, 5), (180, 10)] 

    plt.figure(figsize=(10, 6))
    for mean, variance in params:
        y_values = [normal_density(mean, variance, x) for x in x_values]
        plt.plot(x_values, y_values, label=f'µ={mean}, σ²={variance}')
    
    plt.title('Normal Density Functions')
    plt.xlabel('x')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_normal_density()
```

```python
import numpy as np
from scipy.integrate import quad

def normal_density(mean, variance, x):
    coefficient = 1 / np.sqrt(2 * np.pi * variance)
    exponent = -((x - mean) ** 2) / (2 * variance)
    return coefficient * np.exp(exponent)

def expected_value_uniform(a, b):
    return (a + b) / 2

def expected_value_exponential(rate):
    return 1 / rate

def expected_dosage(mean, variance):
    result, _ = quad(lambda x: 2.38 * x ** 2 * normal_density(mean, variance, x), -np.inf, np.inf)
    return result

# (a) Uniform Distribution Expectation
a = 1
b = 5
print(f"Expected value for uniform distribution on [{a}, {b}]: {expected_value_uniform(a, b):.2f}")

# (b) Exponential Distribution Expectation
rate = 1 / 50
print(f"Expected value for exponential distribution with rate {rate}: {expected_value_exponential(rate):.2f}")

# (c) Average drug dosage for a random male
mean = 171
variance = 7.1
average_dosage = expected_dosage(mean, variance)
print(f"Estimated average drug dosage for a random male: {average_dosage:.2f} mg")
```
